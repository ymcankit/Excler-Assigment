```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
sales_df = pd.read_csv('sales_data_with_discounts.csv')
```

```python
sales_df.head()
```

```python
sales_df.describe()
```

```python
# Identify numerical columns
Numerical_columns = sales_df.select_dtypes(include=['number']).columns
print("Numerical Columns:", Numerical_columns)
```

```python
#Calculate the mean, median, mode, and standard deviation

mean_values = sales_df[Numerical_columns].mean()
median_values = sales_df[Numerical_columns].median()
mode_values = sales_df[Numerical_columns].mode().iloc[0]
std_values = sales_df[Numerical_columns].std()

print("Mean Values:")
print(mean_values)

print("\nMedian Values:")
print(median_values)

print("\nMode Values:")
print(mode_values)

print("\nStandard Deviation Values:")
print(std_values)
```

```python
# Interpretation of Statistics

'''Mean: Indicates the average value in the dataset

   Median: Provides the middle point of the data distribution, useful when the data contains outliers

   Mode: Shows the most frequent occurrence, which can be useful to understand popular discount rates or volume sales

   Standard Deviation: Tells you how spread out the values are from the mean. A higher standard deviation indicates
                      greater variability in sales or pricing

'''
```

```python
#Data Visualization
# Histogram: Histograms show how the data is distributed (e.g., skewness, presence of outliers)
# Boxplot: Boxplots help identify outliers and the interquartile range (
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize histograms for numerical columns
for col in Numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(sales_df[col])
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Visualize boxplots for numerical columns
for col in Numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=sales_df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
# Bar Chart Analysis for Categorical Column

# Identify categorical columns in the dataset
categorical_columns = sales_df.select_dtypes(include=['object']).columns

# Plot bar charts for each categorical column
for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=sales_df, x=col)
    plt.title(f'Bar Chart for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.show()
```

```python
# Insights from Categorical Bar Charts:

# 1. Category Popularity:
#     Identify the most and least frequent categories
#     Example: In "Region," if `North` dominates, most sales occur there

# 2. Distribution Imbalance:
#    Uneven distributions may highlight focus areas or biases
#    Example: A dominant product type like `Electronics` may indicate other types are underrepresented

# 3. Opportunities:
#    Low-frequency categories suggest areas for growth
#    Example: If `South` has fewer sales, focus marketing there

# 4. Data Validation:
#    Spot anomalies or unexpected categories for cleaning
#    Example: Categories like `Unknown` or `Null` need investigation
```

```python
#Standardization of Numerical Variables
```

```python
""" Z-Score : Z-score normalization also known as standardization is a crucial data preprocessing technique in machine learning
and statistics. It is used to transform data into a standard normal distribution ensuring that all features are on
the same scale. This process helps to avoid the dominance of certain features over others due to differences in their
scales which can significantly impact the performance of machine learning models


formula = z=x-mu/sigma


x: Original value
mu: Mean of the column
sigma: Standard deviation of the column

"""
```

```python

# Standardize each numerical column using Z-score formula
sales_df_standardized = sales_df.copy()
for col in Numerical_columns:
    mean = sales_df[col].mean()  # Mean of the column
    std = sales_df[col].std()    # Standard deviation of the column
    sales_df_standardized[col] = (sales_df[col] - mean) / std  # Z-score standardization


# Compare the before and after standardization
print("Before Standardization:")
print(sales_df[Numerical_columns].describe())#

print("\nAfter Standardization:")
print(sales_df_standardized.describe())

```

```python
# Need for converting categorical data into dummy variables (one-hot encoding):
#One-hot encoding is essential for converting categorical data into a form that is interpretable and usable by
#machine learning algorithms without introducing bias or incorrect assumptions
```

```python
# Applying one-hot encoding to the relevant categorical columns
categorical_columns = ['Day', 'City', 'BU', 'Brand', 'Model']
encoded_data = pd.get_dummies(sales_df, columns=categorical_columns, drop_first=False)

# Displaying a portion of the transformed dataset
encoded_data.head()
```
